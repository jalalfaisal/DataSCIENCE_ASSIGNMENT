```python
import pandas as pd

# Create the initial Series
int_series = pd.Series([1, 2, 3], dtype=int)

# Set an element to None
int_series[1] = None

# Print the resulting Series and its data type
print(int_series)
print("dtype:", int_series.dtype)
```

The element at index 1 has been replaced with NaN (Not a Number), indicating a missing value.
The data type of the Series is object, as it now contains mixed types (integers and a special floating-point value).

Numeric Data

```python
fill_with_mean = pd.DataFrame([[-2,0,1],
                               [-1,2,3],
                               [np.nan,4,5],
                               [1,6,7],
                               [2,8,9]])

fill_with_mean
```

```python
# THE MEAN OF THE COLUMN
np.mean(fill_with_mean[0])
```

```python
# FILL WITH THE MEAN
fill_with_mean[0].fillna(np.mean(fill_with_mean[0]),inplace=True)
fill_with_mean
```

```python
# FILL WITH THE MEDIAN
fill_with_median = pd.DataFrame([[-2,0,1],
                               [-1,2,3],
                               [0,np.nan,5],
                               [1,6,7],
                               [2,8,9]])

fill_with_median
```

```python
# THE MEADIAN OF THE SECOND COLUMN
fill_with_median[1].median()
```

```python
#FILLING WITH THE MEDIAN
fill_with_median[1].fillna(fill_with_median[1].median(),inplace=True)
fill_with_median
```

LABEL ENCODING

```python
label = pd.DataFrame([
                      [10,'business class'],
                      [20,'first class'],
                      [30, 'economy class'],
                      [40, 'economy class'],
                      [50, 'economy class'],
                      [60, 'business class']
],columns=['ID','class'])
label
```

```python
#To perform label encoding on the 1st column, we have to first describe a mapping from each class to a number, before replacing
class_labels = {'business class':0,'economy class':1,'first class':2}
label['class'] = label['class'].replace(class_labels)
label
```

ONE HOT ENCODING

```python
one_hot = pd.DataFrame([
                      [10,'business class'],
                      [20,'first class'],
                      [30, 'economy class'],
                      [40, 'economy class'],
                      [50, 'economy class'],
                      [60, 'business class']
],columns=['ID','class'])
one_hot
```

```python
# one hot encoding on the 1st column
one_hot_data = pd.get_dummies(one_hot,columns=['class'])
```

```python
one_hot_data
```
