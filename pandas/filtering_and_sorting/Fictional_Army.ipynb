```python
# Import libraries
import pandas as pd
```

```python
# Create an example dataframe about a fictional army
raw_data = {'regiment': ['Nighthawks', 'Nighthawks', 'Nighthawks', 'Nighthawks', 'Dragoons', 'Dragoons', 'Dragoons', 'Dragoons', 'Scouts', 'Scouts', 'Scouts', 'Scouts'],
            'company': ['1st', '1st', '2nd', '2nd', '1st', '1st', '2nd', '2nd','1st', '1st', '2nd', '2nd'],
            'deaths': [523, 52, 25, 616, 43, 234, 523, 62, 62, 73, 37, 35],
            'battles': [5, 42, 2, 2, 4, 7, 8, 3, 4, 7, 8, 9],
            'size': [1045, 957, 1099, 1400, 1592, 1006, 987, 849, 973, 1005, 1099, 1523],
            'veterans': [1, 5, 62, 26, 73, 37, 949, 48, 48, 435, 63, 345],
            'readiness': [1, 2, 3, 3, 2, 1, 2, 3, 2, 1, 2, 3],
            'armored': [1, 0, 1, 1, 0, 1, 0, 1, 0, 0, 1, 1],
            'deserters': [4, 24, 31, 2, 3, 4, 24, 31, 2, 3, 2, 3],
            'origin': ['Arizona', 'California', 'Texas', 'Florida', 'Maine', 'Iowa', 'Alaska', 'Washington', 'Oregon','teddy','houston','York']}
```

```python
# Create a dataframe and assign it to a variable called army.
army = pd.DataFrame(data=raw_data)
army
```

```python
# Set 'origin' as index
army.set_index('origin', inplace=True)
```

```python
# Print 'veterans' column
print(army["veterans"])
```

```python
# Print 'veterans' and 'deaths' columns
print(army[["veterans","deaths"]])
```

```python
# Print names of all columns
army.columns
```

```python
# Select 'deaths', 'size' and 'deserters' columns from Maine and Alaska
army.loc[["Maine","Alaska"],["deaths","size","deserters"]]
```

```python
# Select the rows 3 to 7 and the columns 3 to 6
army.iloc[2:7,2:6]
```

```python
#  Select every row after the fourth row and all columns
army.iloc[4:, :]
```

```python
#  Select rows where df.deaths is greater than 50
army[army["deaths"] > 50]
```

```python
# Select rows where df.deaths is greater than 500 or less than 50
army[(army["deaths"] > 500) | (army["deaths"] < 50)]
```

```python
#  Select all the regiments not named "Dragoons"
army[army["regiment"] != "Dragoons"]
```

```python
# Select the rows called Texas and Arizona
army.loc[["Texas", "Arizona"], :]
```
